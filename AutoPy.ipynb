### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# https://pyautogui.readthedocs.io/en/latest/mouse.html

# Passo 1: Abrir navegador acessar site
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 0.5

#pyautogui.hotkey("ctrl","t") # Abrir nova aba

pyautogui.press("win") # Pressioa tecla win (Windows)
pyautogui.write("chrome") # Escreve o nome do navegador (Chrome)
pyautogui.press("enter") # Pressiona Enter para abrir

pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga") # Copia link do site
pyautogui.hotkey("ctrl","v") # Cola o link do site no navegador
pyautogui.press("enter") # Pressiona Enter para acessar

# Passo 2: Navegar no computador
time.sleep(3) # Espera o site carregar
pyautogui.click(x=388, y=309, clicks=2)
# Exemplo se quiser clicar 3 vezes com o botão direito do mouse
# pyautogui.click(button='right', clicks=3, interval=0.25)  

# Passo 3: Exportar/fazer Download
time.sleep(1) # Espera o site carregar
pyautogui.click(x=455, y=410) # Clicar no arquivo
time.sleep(0.5) # Espera o site carregar
pyautogui.click(x=1153, y=193) # Clicar nos 3 pontinhos
time.sleep(0.5) # Espera o site carregar
pyautogui.click(x=945, y=624) # Clicar em fazer Donwload
time.sleep(5) # Esperar o DownLoad acontecer
```

```python
# Passo 4: Importar a base de dados
# usar as pd para nomear pandas como pd
import pandas as pd

# tabela = pd.read_csv()
# coloca r antes do caminho para python não errar interpretação de \
tabela = pd.read_excel(r"C:\Users\rafae\Downloads\Vendas - Dez.xlsx") # lê arquivo 
display(tabela) # Mostra tabela
# macete = pd. + press tab para ver o que pode fazer
```

```python
# passo 5: Calcular faturamento
quantidade = tabela["Quantidade"].sum()
faturamento = tabela["Valor Final"].sum()
#print(quantidade)
#print(faturamento)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 6: Enviar e-mail
# Abrir e-mail
pyautogui.hotkey("ctrl","t") # Abre nova aba
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox") # Copia link do e-mail
pyautogui.hotkey("ctrl","v") # Cola link
pyautogui.press("enter")
time.sleep(4) # Espera e-mail carregar

# Clicar no botão escrever
pyautogui.click(x=130, y=210)
time.sleep(2) # Espera caixa do e-mail abrir

# Escrever endereço de e-mail destinatário
pyautogui.write("nobre_17@hotmail.com")
pyautogui.press("tab") # Selecionando destinatário
pyautogui.press("tab") # Passa para campo do assunto

# Escrever assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl","v") # Copia assunto
pyautogui.press("tab") # Selecionando destinatário

# Escrever corpo do e-mail
# inserir f no início do texto para indicar parâmetro no meio do texto
# Usar 3 aspas para escrever pulando linha
# Formatar número(:inserir 2 pontos e depois ,separador de milhar .casa decimal)
# Quantidade de casa decimal + f fica assim :.2f
texto = f"""Prezados, bom dia.

A quantidade de produtos foi de: R$ {quantidade:,.2f}
O faturamento de ontem foi de: R$ {faturamento:,}
  
Att,
Rafael"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl","v")
time.sleep(2) # Espera 2 segundos para ler o e-mail

# Enviar e-mail
pyautogui.click(x=840, y=625)
```
